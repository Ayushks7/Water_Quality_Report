```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.scatterplot(x='ph',y='Hardness',data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['ph'], y=df['Hardness'], color='green')

plt.title(" ph vs Fire hardness")

plt.xlabel("ph")

plt.ylabel("hardness")

plt.show()
```

```python
fire_counts_by_day = df.groupby("ph").size()

plt.figure(figsize=(8,5))

fire_counts_by_day.plot(kind='bar', color='purple')

plt.title("Water quality")

plt.xlabel("ph")

plt.ylabel("Potability")

plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["ph"], bins=20, kde=True, color="skyblue")
plt.xlabel("ph")
plt.ylabel("Hardness")
plt.title("Water Quality Report")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df['ph_range'] = pd.cut(df['ph'], bins=[0, 4, 5.5, 7, 8.5, 10, 14], labels=['0-4', '4-5.5', '5.5-7', '7-8.5', '8.5-10', '10-14'])
ph_avg_hardness = df.groupby("ph_range")["Hardness"].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x="ph_range", y="Hardness", data=ph_avg_hardness, palette="coolwarm")
plt.xlabel("pH Range")
plt.ylabel("Average Hardness")
plt.title("Average Hardness of Water Across pH Ranges")
plt.tight_layout()
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(
    x="Solids",
    y="Turbidity",
    hue="Potability",
    data=df,
    kind="scatter",
    height=6)
plt.xlabel("Solids (ppm)")
plt.ylabel("Turbidity (NTU)")
plt.title("Solids vs Turbidity Colored by Potability")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
selected_features = ['ph', 'Hardness', 'Solids', 'Turbidity', 'Potability']
plt.figure(figsize=(8, 6))
sns.heatmap(df[selected_features].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
